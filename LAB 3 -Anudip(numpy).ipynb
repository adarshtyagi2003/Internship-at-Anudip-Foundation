# Q1. calculate total revenue generated by 2 product categories in a store

```python
import numpy as np 
```

```python
category1_revenue = np.array([500,600,700,550])
category2_revenue = np.array([450,700,800,600])
total_revenue=category1_revenue+category2_revenue
print(total_revenue)
```

# Q2. calculate profit made by company 

```python
revenue=np.array([10000,12000,11000,10500])
expenses=np.array([4000,5000,4500,4800])
print(revenue)
print(expenses)
profit=revenue-expenses
print(profit)
```

# Q3. determine which product in a store are out of stock(quantity is 0)

```python
inventory=np.array([10,0,5,0,20,0])
print("out of stock")
print(inventory[inventory==0])    
```
